<h1 style="color:green">1.Bubble sort(by id)</h1>

Performance: Bubble sort has a worst-case time complexity of O(n²) and is inefficient for large datasets. It compares and swaps adjacent elements repeatedly, and in the worst case, it requires many passes through the list

<h2 style="color:blue">Advantages:</h2>

.Simple to understand and implement.

.Efficient for small datasets or nearly sorted lists.

<h2 style="color:blue">Disadvantages:</h2>

.Very slow for large datasets.

.not efficient for general purpose sorting

<h1 style="color:green">2.Insertion Sort (by Name)</h1>

<h2 style="color:blue">Performance: </h2>Insertion sort also has a worst-case time complexity of O(n²), but it performs better than bubble sort for partially sorted lists because it only shifts the elements when necessary.

<h2 style="color:blue">Advantages:</h2>

.Efficient for small or nearly sorted datasets.

.Stable sort, meaning it preserves the order of equal elements.

<h2 style="color:blue">Disadvantages:</h2>

.Slower than more advanced algorithms (e.g., Merge Sort) for larger datasets.

.Requires shifting elements for each insertion, which can be costly for large lists.

.It performs the same number of comparisons regardless of the order of the elements 

algorithm for insertion sort

.initialization

.key selection

.comparision and shifting

.insertion

<h2 style="color:blue">time complexity :</h2>

best case complexity sorted array O(n)

Average case complexity :mostly equal to worst case complexity O(n2)

space complexity :stable sorting algorithm and therfore space complexity is O(n1)



```python
# Creating student profiless
students =[{"ID": "DBU150000", "Name": "Dawit", "dept": "DS", "cgpa": 3.34},
          {"ID": "DBU1500011", "Name": "Abrham", "dept": "CS", "cgpa": 3.12},
          {"ID": "DBU1500022", "Name": "Soliana", "dept": "DS", "cgpa": 2.74},
          {"ID": "DBU1500043", "Name": "Rahel", "dept": "CS", "cgpa": 3.10},
          {"ID":"DBU1500024", "Name": "Bekele", "dept": "IT", "cgpa": 3.56},
          {"ID": "DBU1500015", "Name": "Helen", "dept": "CS", "cgpa": 3.78},
          {"ID": "DBU1500066", "Name": 'Seniet', 'dept': 'DS', 'cgpa': 3.89},
          {"ID": "DBU1500017", "Name": "Teddy", "dept": "IT", "cgpa": 2.95},
          {"ID": "DBU1500018", "Name": "Mengesh", "dept": "CS", "cgpa": 3.43},
          {"ID": "DBU1500019", "Name": "Yetm", "dept": "IT", "cgpa": 3.23}]

#insertion sort 
def Insertion_Sort_By_Name(students):
    n = len(students)
    for i in range(1 , n):
        key_student = students[i]
        j=i-1
        while j>=0 and students[j]["Name"] > key_student["Name"]:
            students[j + 1]= students[j]
            j-=1
            students[j+1]=key_student
Insertion_Sort_By_Name(students)
print("Sorted alfabetically By Name:")
for i in students:
    print(i) 
```


<h1 style="color:green">3.Selection Sort (by CGPA in Descending Order)</h1>

Performance: Selection sort, like bubble sort and insertion sort, has a time complexity of O(n²). It selects the largest (or smallest) element in the unsorted part of the list and swaps it into the correct position.

<h2 style="color:blue">Advantages:</h2>

.Simple to implement.

.Does not require additional memory like merge sort or quicksort.

<h2 style="color:blue">Disadvantages:</h2>

.Inefficient for large datasets.

<h2 style="color:blue">time complexity:</h2>

worst case-  expressed[big-O]O(n2)

best case - expressed as [big-O]O(n2)

Average case- expressed as [big-theta]O(n2)

```python

# Creating student profiless
students =[{"ID": "DBU150000", "Name": "Dawit", "dept": "DS", "cgpa": 3.34},
          {"ID": "DBU1500011", "Name": "Abrham", "dept": "CS", "cgpa": 3.12},
          {"ID": "DBU1500022", "Name": "Soliana", "dept": "DS", "cgpa": 2.74},
          {"ID": "DBU1500043", "Name": "Rahel", "dept": "CS", "cgpa": 3.10},
          {"ID":"DBU1500024", "Name": "Bekele", "dept": "IT", "cgpa": 3.56},
          {"ID": "DBU1500015", "Name": "Helen", "dept": "CS", "cgpa": 3.78},
          {"ID": "DBU1500066", "Name": 'Seniet', 'dept': 'DS', 'cgpa': 3.89},
          {"ID": "DBU1500017", "Name": "Teddy", "dept": "IT", "cgpa": 2.95},
          {"ID": "DBU1500018", "Name": "Mengesh", "dept": "CS", "cgpa": 3.43},
          {"ID": "DBU1500019", "Name": "Yetm", "dept": "IT", "cgpa": 3.23}]
def Selection_Sort_By_cgpa_Descending(students):
    n = len(students)
    for i in range(n):
        max_index = i
        for j in range(i + 1, n):
            if students[j]["cgpa"] > students[max_index]["cgpa"]:
                max_index = j
        students[i] , students[max_index] = students[max_index] , students[i]
Selection_Sort_By_cgpa_Descending(students)
print("Sorted By cgpa(Decending):")
for i in students:
    print(i) 
```

<h1 style="color:green">Conclusion</h1>
Bubble Sort is very simple but inefficient for large datasets due to its O(n²) time complexity. It works well when the list is nearly sorted or small.

Insertion Sort performs better than Bubble Sort on nearly sorted data, with a best-case time complexity of O(n), but still suffers from inefficiency on large datasets.

Selection Sort is efficient in terms of memory usage (O(1) space), but its O(n²) time complexity makes it unsuitable for large datasets.
